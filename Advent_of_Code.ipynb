## Day 2: I Was Told There Would Be No Math

```python
with open("day_2_input.txt") as file:
  lines = [line.rstrip() for line in file]

print(lines)
```

```python
wrapping_paper_needed = 0
ribbon_needed = 0

for line in lines:
  dims = line.split("x")
  dims = [int(dim) for dim in dims]
  area_A = dims[0] * dims[1]
  area_B = dims[1] * dims[2]
  area_C = dims[2] * dims[0]
  little_extra = min(area_A, area_B, area_C)
  wrapping_paper_one_gift = 2*area_A + 2*area_B + 2*area_C + little_extra
  wrapping_paper_needed += wrapping_paper_one_gift

  smallest_perimeter = 2 * (min(dims) + statistics.median(dims))
  bow = dims[0] * dims[1] * dims[2]
  ribbon_one_gift = smallest_perimeter + bow
  ribbon_needed += ribbon_one_gift

print(f"The elves should order {wrapping_paper_needed} square feet of wrapping paper and {ribbon_needed} feet of ribbon.")
```

```python
# 2015 / Day 5
def contains_three_vowels_aeiou(word):
  vowels = "aeiou"
  n = 0
  for letter in word:
    if letter.lower() in vowels:
      n += 1
  return n >= 3

def contains_double_letter(word):
  double_letter = False
  for i in range(len(word) - 1):
    if word[i].lower() == word[i+1].lower():
      return True
  return False

def not_contains_ab_cd_pq_xy(word):
  flag = True
  if "ab" in word:
    flag = False
  if "cd" in word:
    flag = False
  if "pq" in word:
    flag = False
  if "xy" in word:
    flag = False
  return flag

def is_nice_string(word):
  return (contains_three_vowels_aeiou(word) and
          contains_double_letter(word) and
          not_contains_ab_cd_pq_xy(word))

def contains_pair_twice(word):
  pair_twice = False
  for i in range(len(word) - 3):
    if word[i:i+2].lower() in word[i+2:].lower():
      return True
  return False

def contains_double_letter_with_between(word):
  double_letter_with_between = False
  for i in range(len(word) - 2):
    if word[i].lower() == word[i+2].lower():
      return True
  return False

def is_even_nicer_string(word):
  return (contains_pair_twice(word) and
          contains_double_letter_with_between(word))
```

```python
# 2015 / Day 6
def touple_ints_from_string_with_comma_sep_two_numbers(string):
  result = string.split(",")
  result = [int(num) for num in result]
  return result[0], result[1]

def range_2D_from_2_touples(touple1, touple2):
  result = []
  x1, y1 = touple1[0], touple1[1]
  x2, y2 = touple2[0], touple2[1]
  for x in range(x1, x2+1):
    for y in range(y1, y2+1):
      result.append((x, y))
  return result
```

```python
# Python3 program to implement traveling salesman
# problem using naive approach.
from sys import maxsize
from itertools import permutations
V = 8

# implementation of traveling Salesman Problem
def travellingSalesmanProblem(graph, s):

    # store all vertex apart from source vertex
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)

    # store minimum weight Hamiltonian Cycle
    min_path = maxsize
    max_used_in_result = 0 ##
    used_permutation = () ##
    min_path_wo_return = maxsize ##
    next_permutation=permutations(vertex)
    for i in next_permutation:

        # store current Path weight(cost)
        current_pathweight = 0

        ## store current max section used
        max_used = 0

        ## store current permutation
        current_permutation = i

        # compute current path weight
        k = s
        for j in i:
            current_pathweight += graph[k][j]
            max_used = max(max_used, graph[k][j])
            k = j
        current_pathweight += graph[k][s]
        max_used = max(max_used, graph[k][s])

        ## if new minimum found, update max_used_in_result & used permutation
        ## if current_pathweight < min_path:
        if current_pathweight - max_used < min_path_wo_return:
          max_used_in_result = max_used
          used_permutation = current_permutation
          print(current_pathweight, max_used, current_pathweight - max_used, current_permutation)

        # update minimum
        min_path = min(min_path, current_pathweight)

        ## update minimum w/o returning to start
        min_path_wo_return = min(min_path_wo_return, current_pathweight - max_used)


    return min_path, max_used_in_result, min_path_wo_return, used_permutation ##
```

```python
from sys import maxsize
from itertools import permutations
V = 8

# implementation of INVERSE traveling Salesman Problem
def travellingSalesmanProblemInversed(graph, s):

    # store all vertex apart from source vertex
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)

    # store minimum weight Hamiltonian Cycle
    max_path = -maxsize - 1
    max_used_in_result = 0 ##
    used_permutation = () ##
    max_path_wo_return = -maxsize - 1 ##
    next_permutation=permutations(vertex)
    for i in next_permutation:

        # store current Path weight(cost)
        current_pathweight = 0

        ## store current max section used
        min_used = maxsize

        ## store current permutation
        current_permutation = i

        # compute current path weight
        k = s
        for j in i:
            current_pathweight += graph[k][j]
            min_used = min(min_used, graph[k][j])
            k = j
        current_pathweight += graph[k][s]
        min_used = min(min_used, graph[k][s])

        ## if new maximum found, update max_used_in_result & used permutation
        ## if current_pathweight > min_path:
        if current_pathweight - min_used > max_path_wo_return:
          max_used_in_result = min_used
          used_permutation = current_permutation
          print(current_pathweight, min_used, current_pathweight - min_used, current_permutation)

        # update maximum
        max_path = max(max_path, current_pathweight)

        ## update maximum w/o returning to start
        max_path_wo_return = max(max_path_wo_return, current_pathweight - min_used)


    return max_path, max_used_in_result, max_path_wo_return, used_permutation ##
```

```python
import statistics

def is_triangle(a, b, c):
  minimum = min(a, b, c)
  second = statistics.median([a, b, c])
  maximum = max(a, b, c)
  if (minimum + second) > maximum:
    return 1
  else:
    return 0
```

## Day 4: Security Through Obscurity

```python
def is_real_room(encrypted_name):
  encrypted_name = encrypted_name[:-1]
  checksum = encrypted_name[encrypted_name.find("[")+1:]
  #print(f"Checksum: {checksum}")
  encrypted_name = encrypted_name[:encrypted_name.find("[")]
  sector_ID = int(encrypted_name[encrypted_name.rfind("-")+1:])
  #print(f"Sector ID: {sector_ID}")
  encrypted_name = encrypted_name[:encrypted_name.rfind("-")]
  encrypted_name = encrypted_name.replace("-", "")
  #print(f"Encrypted name: {encrypted_name}")
  sorted_name = "".join(sorted(encrypted_name))
  #print(sorted_name)
  freq = {}
  for l in sorted_name:
    if l in freq:
      freq[l] += 1
    else:
      freq[l] = 1
  #print(freq)
  actual_checksum = ""
  for i in range(5):
    next_letter = max(freq, key=freq.get)
    actual_checksum += next_letter
    del freq[next_letter]
  #print(actual_checksum)
  if actual_checksum == checksum:
    return sector_ID
  else:
    return 0
```

```python
def shift_lower_letter(char, shift):
  numerical = (ord(char) + shift)
  if numerical > 122:
    numerical -= 26
  return chr(numerical)
```

```python
def decrypt_name(encrypted_name):
  encrypted_name = encrypted_name[:encrypted_name.find("[")]
  sector_ID = int(encrypted_name[encrypted_name.rfind("-")+1:])
  #print(f"Sector ID: {sector_ID}")
  shift = sector_ID % 26
  #rint(shift)
  encrypted_name = encrypted_name[:encrypted_name.rfind("-")]
  encrypted_name = encrypted_name.replace("-", " ")
  #print(f"Encrypted name: {encrypted_name}")
  decrypted_name = ""
  for letter in encrypted_name:
    if letter == " ":
      pass
    else:
      letter = shift_lower_letter(letter, shift)
    decrypted_name += letter
  #print(f"Decrypted name: {decrypted_name}")
  return decrypted_name, sector_ID
```

```python
def has_abba(string):
    for i in range(len(string) - 3):
        pair = string[i:i+2]
        if pair[0] != pair[1]:  # Ensure characters in the pair aren't the same
            reverse_pair = pair[::-1]
            if string[i+2:i+4] == reverse_pair:
                return True
    return False
```

```python
def split_and_keep_brackets(string):
    import re
    parts = re.split(r'(\[|\])', string)
    return [part for part in parts if part]
```

```python
print(split_and_keep_brackets(day_7_test1))
print(split_and_keep_brackets(day_7_test2))
print(split_and_keep_brackets(day_7_test3))
print(split_and_keep_brackets(day_7_test4))
```

```python
def supports_tls(string):
    list = split_and_keep_brackets(string)
    brackets = False
    abba = False
    for part in list:
        if part == "[":
            brackets = True
        elif part == "]":
            brackets = False
        elif has_abba(part):
            abba = True
            if brackets:
                return False
    return abba
```

```python
print(supports_tls(day_7_test1))
print(supports_tls(day_7_test2))
print(supports_tls(day_7_test3))
print(supports_tls(day_7_test4))
```

```python
def split_list_into_two(list):
    outside = []
    inside = []
    brackets = False
    for part in list:
        if part == "[":
            brackets = True
        elif part == "]":
            brackets = False
        else:
            if brackets:
                inside.append(part)
            else:
                outside.append(part)
    return inside, outside
```

```python
def supports_ssl(list_1, list_2):
    for string_1 in list_1:
        if len(string_1) >= 3:
            for i in range(len(string_1) - 2):
                sequence = string_1[i:i+3]
                if sequence[0] == sequence[2]: # Ensure outside characters are the same
                    if sequence[0] != sequence[1]:  # Ensure the middle character in the sequence isn't the same
                        bab = sequence[1:] + sequence[1]
                        for string_2 in list_2:
                            if bab in string_2:
                                return True
    return False
```

```python
list_1, list_2 = split_list_into_two(split_and_keep_brackets(day_7_test5))
print(supports_ssl(list_1, list_2))
list_1, list_2 = split_list_into_two(split_and_keep_brackets(day_7_test6))
print(supports_ssl(list_1, list_2))
list_1, list_2 = split_list_into_two(split_and_keep_brackets(day_7_test7))
print(supports_ssl(list_1, list_2))
list_1, list_2 = split_list_into_two(split_and_keep_brackets(day_7_test8))
print(supports_ssl(list_1, list_2))
```
